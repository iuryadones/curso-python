### Modificar valores do Recorte da imagem com condicionais

```python
line = 3
column = 3
```

```python
crop_img = image[0:line + 1, 0:column + 1].copy()
```

```python
cond_img = crop_img <= 143
```

```python
crop_img[cond_img] = 0
```

```python
crop_img[~cond_img] = 255
```
