```python
from skimage import data
```

```python
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
data_img = [d for d in dir(data) if not d.startswith('_')]
```

```python
removes = ['data_dir','expected_warnings', 
           'img_as_bool', 'imread',
           'load','osp', 
           'use_plugin', 'warn', 'lfw_subset', 
           'lbpcascade_frontalface_opencv',
           'lbp_frontal_face_cascade_filename']

for err in removes:
    if err in data_img:
        data_img.remove(err)
```

```python
len(data_img)
```

```python
for name in data_img:
    caller = getattr(data, name)
    image = caller()
    
    if isinstance(image, tuple):
        image = image[0]
    
    plt.figure()
    plt.title(name)
    if image.ndim == 2:
        plt.imshow(image, cmap=plt.cm.gray)
    else:
        plt.imshow(image)
plt.show()
```

```python
fig, axes = plt.subplots(4, 5, figsize=(15, 7))
ax = axes.ravel()
images = data.lfw_subset()
for i in range(20):
    ax[i].imshow(images[90+i], cmap=plt.cm.gray)
    ax[i].axis('off')
fig.tight_layout()
plt.show()
```
