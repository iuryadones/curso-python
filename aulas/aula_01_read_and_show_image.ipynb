## Image Load 

```python
from skimage import data
```

```python
image = data.coins()
```

```python
type(image)
```

### Image Matrix

```python
image
```

### Image Show 

```python
from skimage import io
```

```python
io.imshow(image)
```

## Sobel edge detection algorithm

```python
from skimage.filters import sobel
```

```python
edges = sobel(image)
```

```python
type(edges)
```

### Edges Show 

```python
io.imshow(edges)
```

```python
edges
```
