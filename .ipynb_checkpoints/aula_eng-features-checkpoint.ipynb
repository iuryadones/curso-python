```python
from skimage import data, io, filters
from skimage.morphology import disk
from skimage import morphology
```

```python
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
image = data.coins()
```

```python
edges = filters.sobel(image)
```

```python
io.imshow(image)
```

```python
io.imshow(edges)
```

```python
mask = image - edges
```

```python
io.imshow(mask)
```

```python
median_selem = filters.median(image, morphology.square(5))
```

```python
gaussian = filters.gaussian(image, sigma=1)
```

```python
io.imshow(gaussian)
```

```python
feature = filters.gaussian(image, sigma=1)
feature = filters.median(feature, disk(3))
feature = filters.median(feature, disk(5))
feature = filters.sobel(feature)
```

```python
io.imshow(feature)
```

```python
io.imshow(edges)
```

```python
feature.max()
```

```python
feature.min()
```

```python
feature[feature < feature.max()/2] = 0
```

```python
feature[feature >= feature.max()/2] = 1
```

```python
io.imshow(feature)
```
